## Imports

```python
import pandas as pd
import seaborn as sn
from sklearn import metrics
```

## Load data and prediction

```python
df = pd.read_csv('test.csv')
_pred = pd.read_csv('prediction.csv')
df['pred'] = _pred['label']
df['pred_score'] = _pred['score']
```

```python
df
```

## Analysis

```python
print(metrics.classification_report(df['label'], df['pred']))
```

```python
metrics.confusion_matrix(df['label'], df['pred'])
```

```python
confusion_matrix = pd.crosstab(df['label'], df['pred'], rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True, fmt="d")
```

```python
pd.set_option('display.max_colwidth', -1)
df[df['label'] != df['pred']].sort_values(['pred_score'])
```
